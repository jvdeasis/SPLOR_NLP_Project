```python
import pandas as pd
#import pdfplumber
import docx 
import spacy
from spacy import displacy
import re
import os
```

```python
nlp = spacy.load("en_core_web_md")
```

```python
def getText(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return '\n'.join(fullText)
```

```python
directory = 'C:\\Users\\jonat\\OneDrive\\Desktop\\SPLOR Project\\_ Content from Students'

def list_files(dir):                                                                                                  
    r = []                                                                                                            
    subdirs = [x[0] for x in os.walk(dir)]                                                                            
    for subdir in subdirs:                                                                                            
        files = os.walk(subdir).__next__()[2]                                                                             
        if (len(files) > 0):                                                                                          
            for file in files:                                                                                        
                r.append(subdir + "/" + file)                                                                         
    return r       
```

```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO

def pdf_txt(path):
    rsrcmgr = PDFResourceManager()
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    fp = open(path, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0
    caching = True
    pagenos=set()

    for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages, password=password,caching=caching, check_extractable=True):
        interpreter.process_page(page)

    text = retstr.getvalue()

    fp.close()
    device.close()
    retstr.close()
    return text
```

```python
name_pattern = 'Content from Students\\\(.*)/'
file_pattern = '/(.*)\.'
type_pattern = '\.(.*)'
```

```python
name_list = []

for indiv_files in files:
    name_list.append(re.findall(name_pattern,indiv_files))

file_name =[]

for indiv_files in files:
    file_name.append(re.findall(file_pattern,indiv_files))

file_type =[]

for indiv_files in files:
    file_type.append(re.findall(type_pattern,indiv_files))
```

```python
df = pd.DataFrame({'name': name_list, 'file': file_name,'doc_type': file_type, 'file_path': files})
```

```python
body =[]
counter = 0

for docs in df.file_path:
    if df.doc_type[counter] == ['docx']:
        body.append(getText(docs))
    elif df.doc_type[counter] == ['pdf']:
        body.append(pdf_txt(docs))
    else:
        body.append('')
    counter +=1   
```

```python
df['body'] = body
df
```

```python
pattern1 = '\\n(.*):'
pattern2 = ': (.*)\\n'
pattern3 = '\n(.*)\n'
pattern4 = '\\n(.*)-'   
```

```python
p1_loc = []
p2_loc = []
p3_loc = []
p4_loc = []

for text in df.body:
    p1_loc.append(re.findall(pattern1,text))
    
df['p1'] = p1_loc

for text in df.body:
    p2_loc.append(re.findall(pattern2,text))

df['p2'] = p2_loc    
    
for text in df.body:
    p3_loc.append(re.findall(pattern3,text))

df['p3'] = p3_loc    
    
for text in df.body:
    p4_loc.append(re.findall(pattern4,text))    
    
df['p4'] = p4_loc    
```
